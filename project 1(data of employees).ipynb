```python
with open ("employees.csv","w",newline="") as f:
    writer= csv.DictWriter(f,fieldnames=["id","name","email","gender","active"])
    writer.writeheader()
    for e in employees["data"]:
        id_=e["id"]
        name_=e["name"]
        email_=e["email"]
        gender_=e["gender"]
        active_=1 if e["status"]=="active" else 0
        writer.writerow(
            {
              "id": id_,
              "name":name_,
              "email":email_,
              "gender":gender_,
              "active":active_ 
            }
        )
```
